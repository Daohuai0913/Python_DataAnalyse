```python
import matplotlib_inline
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline


def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(7):
        plt.plot(x, np.sin(x + 1 * 5) * (7 - i) * flip)


sinplot()
```

```python
sns.set()
sinplot()
```
